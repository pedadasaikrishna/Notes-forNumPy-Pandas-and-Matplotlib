### **Definition of `numpy.linspace()`**

`numpy.linspace()` is a function in the **NumPy** library that returns **evenly spaced numbers** over a specified interval. It is particularly useful when you need a specific number of points between a start and stop value.

---

### **Syntax**:
```python
numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
```

### **Parameters**:
- **`start`**: The starting value of the sequence.
- **`stop`**: The end value of the sequence.
- **`num`**: The number of evenly spaced values to generate (default is `50`).
- **`endpoint`**: If `True` (default), `stop` is the last value in the sequence. If `False`, the sequence will not include the stop value.
- **`retstep`**: If `True`, it returns a tuple `(samples, step)`, where `step` is the spacing between values.
- **`dtype`**: The data type of the output array. If not specified, it is inferred.
- **`axis`**: The axis in the result along which the values are stored.

---

### **Return Value**:
- A **NumPy array** containing the specified number of evenly spaced values.

---


````python
numbers = np.linspace(0, 10, 5)
print(numbers)
plt.plot(numbers)
plt.show()
````

````python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a 1x2 grid of subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot on the first subplot
axes[0].plot(x, y1, color='blue')
axes[0].set_title("Sine Wave")

# Plot on the second subplot
axes[1].plot(x, y2, color='red')
axes[1].set_title("Cosine Wave")

# Adjust layout
plt.tight_layout()
plt.show()
````

````python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

# Plot data on each subplot
axes[0].plot(x, y1, label="sin(x)", color="blue")
axes[1].plot(x, y2, label="cos(x)", color="green")

# Set titles and legends
axes[0].set_title("Sine Wave")
axes[1].set_title("Cosine Wave")
axes[0].legend()
axes[1].legend()

plt.tight_layout()
plt.show()
````

````python
from matplotlib.gridspec import GridSpec

# Create a figure with GridSpec
fig = plt.figure(figsize=(10, 6))
gs = GridSpec(2, 2, figure=fig)

# Create subplots
ax1 = fig.add_subplot(gs[0, 0])  # Top-left
ax2 = fig.add_subplot(gs[0, 1])  # Top-right
ax3 = fig.add_subplot(gs[1, :])  # Bottom (spanning two columns)

# Plot data
ax1.plot(x, y1, color='blue')
ax1.set_title("Sine Wave")

ax2.plot(x, y2, color='red')
ax2.set_title("Cosine Wave")

ax3.plot(x, y1 + y2, color='purple')
ax3.set_title("Sine + Cosine")

plt.tight_layout()
plt.show()
````

#viridis cmap


````python
# Generate random data for the scatter plot
x = np.random.rand(50)  # 50 random numbers between 0 and 1 for x-axis
y = np.random.rand(50)  # 50 random numbers between 0 and 1 for y-axis
sizes = np.random.rand(50) * 100  # Random sizes for each point (scaled up to 100)
colors = np.random.rand(50)  # Random color values for each point

# Create a scatter plot
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=0.5)

# Add title and labels
plt.title("Random Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

# Show the plot
plt.colorbar()  # Show color scale
plt.show()
````

````python
# Data
x = np.array([1, 2, 3, 4])
vars = np.array(['apple', 'orange', 'banana', 'Grape'])
e = np.array([0.3, 0, 0, 0])
c = np.array(['r', 'g', 'b', 'y'])

# Create the pie chart
plt.pie(x, labels=vars, explode=e, colors=c, autopct='%1.1f%%', shadow=True)

# Title and enhancements
plt.title("Fruit Distribution", fontsize=14, fontweight='bold', color='darkblue')

# Adjust the position and style of the legend
plt.legend(title='Fruits', loc='upper left', fontsize=10, bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
````
